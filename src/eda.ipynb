```python
import altair as alt

# Save a vega-lite spec and a PNG blob for each plot in the notebook
alt.renderers.enable('mimetype')
# Handle large data sets without embedding them in the notebook
alt.data_transformers.enable('data_server')
```

```python
import pandas as pd

red_wine = pd.read_csv('data/winequality-red.csv', sep = ';')
red_wine['type'] = 'red'
red_wine
```

```python
white_wine = pd.read_csv('data/winequality-white.csv', sep = ';')
white_wine['type'] = 'white'
white_wine
```

```python
df = pd.concat([red_wine, white_wine], axis = 0)
df
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
# for pairwise relationship, might be overplotting

numeric_cols = list(df.select_dtypes(exclude=object).columns)

(
    alt.Chart(df)
    .mark_point(opacity=0.2, size=1, color="tan")
    .encode(
        alt.X(alt.repeat("column"), type="quantitative", scale=alt.Scale(zero=False)),
        alt.Y(alt.repeat("row"), type="quantitative", scale=alt.Scale(zero=False)),
    )
    .properties(height=100, width=100)
    .repeat(row=numeric_cols, column=numeric_cols)
    .configure_axis(titleFontSize=10, labelFontSize=1)
)
```

```python
(
    alt.Chart(df)
    .mark_rect()
    .encode(
        alt.X(
            alt.repeat("column"),
            type="quantitative",
            scale=alt.Scale(zero=False),
            bin=alt.Bin(maxbins=25),
        ),
        alt.Y(
            alt.repeat("row"),
            type="quantitative",
            scale=alt.Scale(zero=False),
            bin=alt.Bin(maxbins=25),
        ),
        alt.Color("count()", scale=alt.Scale(scheme="blues")),
    )
    .properties(height=100, width=100)
    .repeat(row=numeric_cols, column=numeric_cols)
).configure_axis(titleFontSize=10, labelFontSize=1)
```

```python
(
    alt.Chart(df)
    .mark_bar(color = 'tan')
    .encode(
        alt.X(alt.repeat(), type="quantitative", bin=alt.Bin(maxbins=44)),
        alt.Y("count()"),
    )
    .properties(height=100, width=300)
    .repeat(repeat=numeric_cols, columns=3)
)
```
